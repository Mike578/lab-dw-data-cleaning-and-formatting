The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they only contain valid data.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

- *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*

```python
df.head()
```

```python
#Standardize Gender column to 'M' or 'F'

# Step 1: standardize to lowered unspaced string values

df['gender'] = df['gender'].astype(str).str.strip().str.lower()
print(df['gender'].unique())
```

```python
# Step 2: standardize the lowered/stripped values to 'M' and 'F'
df['gender'] = df['gender'].replace('f','F').replace('m', 'M').replace('femal', 'F').replace('female', 'F').replace('m','M').replace('male','M')
print(df['gender'].unique())
```

```python
# State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"
print(df['state'].unique())
print('\n')

print()


df['state'] = df['state'].replace({'Cali':'California', 'AZ':'Arizona', 'WA': 'Washington'})
print(df['state'].unique())
```

```python
# In education, "Bachelors" could be replaced by "Bachelor"
df.head()

df['education'] = df['education'].replace({'Bachelors': 'Bachelor'})
print(df['education'].unique())
```

```python
# In Customer Lifetime Value, delete the % character
df.head()

# NOT WORKING: df[customer_lifetime_value] = df[customer_lifetime_value].strip(%)
# df['customer_lifetime_value'] = df['customer_lifetime_value'].str.replace('%', '')

# Convert 'customer_lifetime_value' column to string data type
df['customer_lifetime_value'] = df['customer_lifetime_value'].astype(str)

# Remove '%' character from 'customer_lifetime_value' column
df['customer_lifetime_value'] = df['customer_lifetime_value'].str.replace('%', '')

# Convert the column to numeric if needed
df['customer_lifetime_value'] = pd.to_numeric(df['customer_lifetime_value'], errors='coerce')

df.sample(5) #returns randomly 5 rows of the dataframe
```
